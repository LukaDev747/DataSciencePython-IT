```python
import pandas as pd
import matplotlib.pyplot as plt

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

rating_counts = ratings['movieId'].value_counts().head()

top_movies = pd.DataFrame({
    'movieId': rating_counts.index,
    'rating_count': rating_counts.values
}).merge(movies[['movieId', 'title']], on='movieId')

plt.figure(figsize=(10, 6))
bars = plt.barh(top_movies['title'], top_movies['rating_count'], color='orange')
plt.xlabel('Liczba ocen')
plt.title('5 najczęściej ocenianych filmów')

for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, 
             f' {int(width)}', 
             ha='left', va='center')
plt.show()
```
