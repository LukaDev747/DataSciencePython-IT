```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv("GlobalLandTemperaturesByMajorCity.csv")
london = df[df["City"] == "London"].copy()
london["dt"] = pd.to_datetime(london["dt"])

london = london[london["dt"].dt.year >= 1970 & (london["dt"].dt.year <= 2020)]

london = london.dropna(subset=["AverageTemperature"])

london["Year"] = london["dt"].dt.year

annual = london.groupby("Year")["AverageTemperature"].mean().reset_index()
annual.head()
```

```python
plt.figure(figsize=(14,6))
sns.regplot(data=annual,x="Year",y="AverageTemperature",scatter_kws={"s":30})
plt.title("rednia temperatura w Londynie 1970-2020")
plt.xlabel("Rok")
plt.ylabel("Temperatura")
plt.show()
```
